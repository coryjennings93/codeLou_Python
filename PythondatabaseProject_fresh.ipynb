```python
# Got rid of properties with 0 sq ft and renamed column names and formated the price column 

query_format_columns = """
SELECT 
  street AS Street, city AS City, zip AS Zip, state AS State, beds AS "Bedroom(s)", baths AS "Bathroom(s)", sq__ft AS "SQ FT", type AS Type, printf("$%,d", price) AS "Price of House"
FROM CLproject
WHERE sq__ft > 0
;
"""


run_query(query_format_columns)
```

```python
#create new table to store last query in new table
Create_TableName = "CLprojects_cleaned"

Create_SQL = f"""
create table if not exists {Create_TableName} as
SELECT 
  street AS Street, city AS City, zip AS Zip, state AS State, beds AS "Bedroom(s)", baths AS "Bathroom(s)", sq__ft AS "SQ FT", type AS Type, price AS "Price of House"
FROM CLproject
WHERE sq__ft > 0
;
"""

cursor.execute(Create_SQL)
```

```python
# Aggregate data getting average cost of house by zip
query_avg_price = """
SELECT 
 zip
,ROUND(avg("Price of House"), 2) AS "Avg Cost"
,ROUND(sum("Price of House"), 2) AS "Total Spent"
,COUNT(*) as "Total Purchases"
FROM CLprojects_cleaned
GROUP BY zip

union all

select 
 '(Total)'
,ROUND(avg("Price of House"), 2) AS "Avg Cost"
,ROUND(sum("Price of House"), 2) AS "Total Spent"
,COUNT(*) as "Total Purchases"
from CLprojects_cleaned

ORDER BY "Total Purchases" DESC;


"""

run_query(query_avg_price)
```

```python
# Aggregate data getting average cost of house by number of bedrooms
query_avg_price_per_bed = """
SELECT 
 "Bedroom(s)"
,ROUND(avg("Price of House"), 2) AS "Avg Cost"
,ROUND(sum("Price of House"), 2) AS "Total Spent"
,COUNT(*) AS "Total Purchases"
,MIN("Price of House") AS "Lowest Price"
,MAX("Price of House") AS "Highest Price"
FROM CLprojects_cleaned
GROUP BY "Bedroom(s)"

union all

select 
 '(Total)'
,ROUND(avg("Price of House"), 2) AS "Avg Cost"
,ROUND(sum("Price of House"), 2) AS "Total Spent"
,COUNT(*) as "Total Purchases"
,MIN("Price of House") AS "Lowest Price"
,MAX("Price of House") AS "Highest Price"
from CLprojects_cleaned

ORDER BY "Bedroom(s)" DESC;


"""

run_query(query_avg_price_per_bed)
```
